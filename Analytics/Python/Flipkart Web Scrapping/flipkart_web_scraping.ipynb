```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import sys
import time
```

```python
url = "https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}"
```

```python
titleList = []
priceList = []
ratingList = []
featureList = []
ratingsCountList = []
reviewsCountList = []
```

```python
try:
  for i in range(1,4):
    print(url.format(i))
    req = requests.get(url.format(i))
    soup = BeautifulSoup(req.text, 'html.parser')
    itemsTable = soup.find('div', {'class':'_1YokD2 _3Mn1Gg'})
    cells = itemsTable.find_all('div', {'class':'_3pLy-c row'})
    
    for cell in cells:
    
        title = cell.find('div', {'class':'_4rR01T'})
        price = cell.find('div', {'class':'_30jeq3 _1_WHN1'})
        rating = cell.find('div', {'class':'_3LWZlK'})
        feature = cell.find('ul', {'class':'_1xgFaf'})
        ratingsCount = cell.find('span', {'class':'_2_R_DZ'})
    
        value = title.text if title.text != "" else ""
        titleList.append(value)

        value = price.text if price.text != "" else ""
        priceList.append(value)

        value = rating.text if rating.text != "" else ""
        ratingList.append(value)
        
        value = [] 
        if feature.text != "":
            for i in feature.text.split("|"):
                i = i.strip()
                value.append(i)
        featureList.append(value)

        value = []
        if ratingsCount.text != "":
            for i in ratingsCount.text.replace("Ratings","").replace("Reviews","").split("&"):
                value.append(int(i.replace(",","")))
        value0 = value[0] if len(value)==2 else 0
        value1 = value[1] if len(value)==2 else 0
        ratingsCountList.append(value0)
        reviewsCountList.append(value1)
        
except Exception as e:
  error_type, error_info, error_obj = sys.exec_info
  print("error in url", url)
  print(error_type, "line no:", error_info)
```

```python
df=pd.DataFrame({'title':titleList,'price':priceList, 'rating':ratingList, 'features':featureList,
                 'ratingCount':ratingsCountList, 'reviewsCount':reviewsCountList})
df
```
